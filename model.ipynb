# Text preprocessing (NLP)

```python
df.info()
```

```python
text_cols = [
    'title',
    'company_profile',
    'description',
    'requirements',
    'benefits'
]

df['combined_text'] = (
    df[text_cols]
    .fillna('')
    .agg(' '.join, axis=1)
)
```

```python
#Filling missing values with '', so that signal is not missed.

for cols in text_cols:
    df[cols] = df[cols].fillna('')
```

```python
# Basic Text cleaning
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r'http\S+', '', text) #remove URLs
    text = re.sub(r'[^a-z\s]', ' ', text) #remove punctuations/numbers
    text = re.sub(r'\s+', ' ', text).strip() 
    return text
for cols in text_cols:
    df[cols] = df[cols].apply(clean_text) 
```

```python
#combine text into one column
df['combined_cols'] = (df['title'] + ' ' + df['company_profile'] + ' ' + df['description'] + ' ' + df['requirements'] + ' ' + df['benefits'])
```

```python
df['industry_missing_flag'] = df['industry_grouped'].isna().astype(int)
df['required_experience_missing_flag'] = df['required_experience'].isna().astype(int)
df['required_education_missing_flag'] = df['required_education'].isna().astype(int)
```

## NLP Feature extraction

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    max_features=20000, #controls dimensionality
    ngram_range=(1, 2), #captures phrases
    min_df=5, #ignore rare noise
    stop_words='english' 
)
```

```python
structured_cols = [
    'telecommuting',
    'has_company_logo',
    'has_questions',

    'salary_fraud_signal_count',

    'industry_missing_flag',
    'required_experience_missing_flag',
    'required_education_missing_flag'
]
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
preprocessor = ColumnTransformer(transformers=[
    ('text', tfidf, 'combined_text'),
    ('num', 'passthrough', structured_cols)
])
```

```python
log_reg = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(
        max_iter=1000,
        class_weight='balanced'
    ))
])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, roc_auc_score, average_precision_score

X = df[['combined_text'] + structured_cols]
y = df['fraudulent']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

log_reg.fit(X_train, y_train)

y_pred_logreg = log_reg.predict(X_test)
y_prob_logreg = log_reg.predict_proba(X_test)[:,1]

print(classification_report(y_test, y_pred_logreg))
print("Roc-AUC : ", roc_auc_score(y_test, y_prob_logreg))

print("PR-Score: ", average_precision_score(y_test, y_prob_logreg))
```

From the above confusion matrix, we got precision 1.00 for 0 (No fraudlent). May be overfitting

```python
train_auc = roc_auc_score(y_train, log_reg.predict_proba(X_train)[:,1])
test_auc = roc_auc_score(y_test, log_reg.predict_proba(X_test)[:,1])

train_auc, test_auc
```

If the gap < 0.02 -> Not overfitting

If the gap > 0.05 -> Overfitting

Ours is not overfitting...

Gmail spam filters often reach ~0.999 AUC. Text problems are very different from pure tabular ML. 

Reasons:
* Vocabulary creates high-dimensional sparse space
* Certain words or phrases can be near-perfect indicators
* Fraud posts often reuse templates

## Let's now use smote and see the accuracy

```python
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline

log_reg_smote = ImbPipeline(steps=[
    ('preprocessor', preprocessor),
    ('smote', SMOTE(random_state=42)),
    ('classifier', LogisticRegression(max_iter=1000, random_state=42))
])

X = df[['combined_text'] + structured_cols]
y = df['fraudulent']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

log_reg_smote.fit(X_train, y_train)

y_pred_logreg_smote = log_reg_smote.predict(X_test)
y_prob_logreg_smote = log_reg_smote.predict_proba(X_test)[:,1]

print(classification_report(y_test, y_pred_logreg_smote))
print("Roc-AUC : ", roc_auc_score(y_test, y_prob_logreg_smote))

print("PR-Score: ", average_precision_score(y_test, y_prob_logreg_smote))
```

Precision for 1(Fake jobs) detection increase -> good sign

### Our dataset is non linear, so maybe logistic regression is overfitting because it is a linear model.

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

preprocessor = ColumnTransformer(transformers=[
    ('text', tfidf, 'combined_text'),
    ('num', 'passthrough', structured_cols)
])

random_forest = ImbPipeline(steps=[
    ('preprocessor', preprocessor),
    ('smote', SMOTE(random_state=42)),
    ('classifier', RandomForestClassifier(
        random_state=42,
        n_estimators=200,
        max_depth=None,
        class_weight='balanced',
        n_jobs=-1
    ))
])

X = df[['combined_text'] + structured_cols]
y = df['fraudulent']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

random_forest.fit(X_train, y_train)

y_pred_rf = random_forest.predict(X_test)
y_prob_rf = random_forest.predict_proba(X_test)[:,-1]

print(classification_report(y_test, y_pred_rf))
print("Roc-AUC: ", roc_auc_score(y_test, y_prob_rf))

print("PR-Score: ", average_precision_score(y_test, y_prob_rf))
```

If the cost of missing a fake job is high (scammers exploiting the platform), go with Logistic Regression + SMOTE → higher recall.

If the cost of flagging legitimate jobs is high, and you can afford to miss some fakes, go with Random Forest/XGBoost → higher precision.

## Moving to BERT

#### Transformer based NLP model

```python
#DistilBERT

import torch
from transformers import(
    DistilBertTokenizerFast,                #Converts text → tokens → token IDs → tensors
    DistilBertForSequenceClassification,    #Classification head → predicts a label (0 / 1)
    Trainer,                                #high-level API from Hugging Face that handles the entire training loop for transformer models.
    TrainingArguments                       #configuration class that tells the Trainer how to train your model.
)
```

```python
print(df.columns)
```

```python
df["combined_text"] = (
    df["title"].fillna("") + " " +
    df["company_profile"].fillna("") + " " +
    df["description"].fillna("") + " " +
    df["requirements"].fillna("") + " " +
    df["benefits"].fillna("")
)
```

```python
df["combined_text"] = df["combined_text"].astype(str)
```

```python
dup_count = df.duplicated(subset=["combined_text"]).sum()
print("Duplicate combined_text:", dup_count)
```

```python
df = df.drop_duplicates(subset=["combined_text"])
```

```python
#We only need text + label.

text = df['combined_text'].tolist()
labels = df['fraudulent'].tolist()
```

```python
print(df.columns)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(text, labels, stratify=labels, random_state=42, test_size=0.2)

#Load tokenizer
tokenizer = DistilBertTokenizerFast.from_pretrained("distilbert-base-uncased")

#Tokenize text
def tokenize(texts):
    return tokenizer(
        texts,
        padding = True,
        truncation = True,
        max_length = 256
    )

train_encodings = tokenize(X_train)
test_encodings = tokenize(X_test)
```

```python
#Create dataset object
#This class converts your tokenized text + labels into a format that PyTorch + HuggingFace Trainer can understand.

class JobDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item["labels"] = torch.tensor(self.labels[idx])
        return item
    
    def __len__(self):
        return len(self.labels)
    
train_dataset = JobDataset(train_encodings, y_train)
test_dataset = JobDataset(test_encodings, y_test)
```

```python
#Load DistilBERT model

distilBERT = DistilBertForSequenceClassification.from_pretrained(
    "distilbert-base-uncased",
    num_labels = 2
)
```

```python
#Handle class imbalance

from sklearn.utils.class_weight import compute_class_weight
import numpy as np

class_weight = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_train),
    y=y_train
)

class_weight = torch.tensor(class_weight, dtype = torch.float)
```

```python
loss_fn = torch.nn.CrossEntropyLoss(weight=class_weight)
```

```python
class FraudTrainer(Trainer):
    def compute_loss(
        self,
        model,
        inputs,
        return_outputs=False,
        num_items_in_batch=None   # ✅ REQUIRED for newer versions
    ):
        labels = inputs.get("labels")

        outputs = model(**inputs)
        logits = outputs.logits

        loss_fct = torch.nn.CrossEntropyLoss(
            weight=class_weight.to(logits.device)
        )

        loss = loss_fct(logits, labels)

        return (loss, outputs) if return_outputs else loss
```

```python
#Training configuration

training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=3,
    weight_decay=0.01,
    logging_dir="./logs",
    load_best_model_at_end=True,
    metric_for_best_model="eval_loss"
)
```

```python
import transformers
print(transformers.__version__)
```

```python
trainer = FraudTrainer(
    model=distilBERT,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset
)

trainer.train()
```

```python
preds = trainer.predict(test_dataset)
y_probs = torch.softmax(torch.tensor(preds.predictions), dim=1)[:,1].numpy()
y_pred = np.argmax(preds.predictions, axis=1)
```

```python
from sklearn.metrics import classification_report, roc_auc_score, average_precision_score

print(classification_report(y_test, y_pred))
print("ROC-AUC: ", roc_auc_score(y_test, y_probs))
```

## Test the model

```python
import torch
from transformers import DistilBertForSequenceClassification, DistilBertTokenizerFast
import torch.nn.functional as F

device = torch.device("mps")

model = DistilBertForSequenceClassification.from_pretrained(
    "./fraud_distilbert_model"
).to(device)

tokenizer = DistilBertTokenizerFast.from_pretrained(
    "./fraud_distilbert_model"
)

model.eval()

text = "Pay registration fee to apply for this job"

inputs = tokenizer(text, return_tensors="pt")
inputs = {k: v.to(device) for k, v in inputs.items()}

with torch.no_grad():
    outputs = model(**inputs)

probs = F.softmax(outputs.logits, dim=1)
print(probs)

pred = torch.argmax(probs, dim=1).item()
confidence = probs[0][pred].item() * 100

label = "Fraud" if pred == 1 else "Not Fraud"
print("The job posting is:", label)
print(f"Confidence: {confidence:.2f}%")
```
